Создадим матрицу с шестью векторами признаков, а также столбец с правильной бинарной классификацией этих объектов (столбец правильный ответов).

````python
training_inputs = np.array ([
    [0,0,1],
    [1,0,1],
    [1,0,0],
    [0,1,1],
    [0,1,0],
    [1,1,1]
])

training_outputs = np.array([[0,2,1,0,0,1]]).T # метод T транспонирует матрицу: превратит строку в столбец правильных ответов
````

````python
training_outputs
````

Функция ```np.random.random((M,N))``` возвращает матрицу случайных чисел из диапазона [0, 1), состояющую из M столбцов и N строк. Если надо получить вектор (по сути одну строку маттрицы), то во второй арргумент N мы можем передать 1.

````python
weights = np.random.random((3,1)) # создаем вектор весов, для начала со случайными значениями

print("Веса до обучения (случайные):")
print(weights)
````

Создадим функцию активации (для этого будем использовать сигмоиду):

````python
def sigmoid(x):
  return 1/(1 + np.exp(-x))
````

Запустим цикл обучения:

````python
for i in range(1000):
   outputs = sigmoid(np.dot(training_inputs, weights)) # вычисляем выходное значение, для простоты тут не используем смещение, только веса
   errors = training_outputs - outputs # вычисляем значения ошибок для каждого вектора в выборке, для простоты берем разность между вычисленным и целевым значением
   adjustments = np.dot(training_inputs.T, errors * outputs) # создаем набор корректировок для каждого веса
   weights = weights + adjustments # применяем корректировки

print("Веса после обучения (скорректированные):")
print(weights)
````

Проверим какие результаты вычислит нейросеть для тестовой выборки:

````python
print(sigmoid(np.dot(np.array([[1,1,0],[0,1,0]]), weights)))
````
